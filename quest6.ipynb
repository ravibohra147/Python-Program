```python
import matplotlib.pyplot as plt

import yfinance as yf

# Extract stock data for Tesla
gme_data = yf.download("TSLA", period="max")

# Reset the index
gme_data.reset_index(inplace=True)
# Assuming make_graph is the function to plot the graph and tesla_data is the dataframe containing Tesla stock data
def make_graph(x, y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()
    
# Assuming make_graph is the function to plot the graph and gme_data is the dataframe containing GameStop stock data
make_graph(gme_data['Date'], gme_data['Close'], title='GameStop Stock Data')
```
