```python
import matplotlib.pyplot as plt

import yfinance as yf

# Extract stock data for Tesla
tesla_data = yf.download("TSLA", period="max")

# Reset the index
tesla_data.reset_index(inplace=True)
# Assuming make_graph is the function to plot the graph and tesla_data is the dataframe containing Tesla stock data
def make_graph(x, y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()

# Plotting Tesla stock graph with title
make_graph(tesla_data['Date'], tesla_data['Close'], title='Tesla Stock Data')
```
